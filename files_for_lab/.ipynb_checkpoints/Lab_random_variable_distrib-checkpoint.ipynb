```python
#0 Read in the data

import pandas as pd

data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
#1 Get the numerical variables from our dataset.

data.dtypes

# Numerical variables:
# Customer Lifetime Value          float64
# Income                             int64
# Monthly Premium Auto               int64
# Months Since Last Claim            int64
# Months Since Policy Inception      int64
# Number of Open Complaints          int64
# Number of Policies                 int64
# Total Claim Amount               float64

numerical = data.select_dtypes(include='number')
numerical.dtypes
```

```python
#2 Check using a distribution plot if the variables fit the theoretical normal 
# or exponential distribution.

import seaborn as sns
import matplotlib.pyplot as plt

def plots():
    for column in numerical.columns:
        sns.displot(numerical[column], bins = 30)
        plt.show()
        
plots()

# I do not see a real normal distirbution 
# However, there are a few clear exponential ones: 
#    - Customer Lifetime Value
#    - Monthly Premium Auto
#    - Number of Open Complaints
#    - Number of Policies
#    - Total Claim Amount
```

```python
#3 Check if any of the transformations (log-transform, etc.) we have seen up to this point 
# changes the result.

# We already know that StandardScaler and MinMaxScaler do not change the original shape of the data - no point in applying these & re-plotting
# But we can apply logn or log10 transformations alternatively

import numpy as np

def log_transfom_clean(x):
    if x!=0:
        return np.log(x)
    else:
        return x # We are returning NaNs so that we can replace them later

for column in numerical.columns:         
    numerical[column] = numerical[column].apply(log_transfom_clean)
```

```python
#3 continued
# To the plots again

def plots():
    for column in numerical.columns:
        sns.displot(numerical[column], bins = 30)
        plt.show()
        
plots()
```

```python
#3 continued
# So after the logn transformation (0's remained 0's), there are changes in the shape:

#    - Customer Lifetime Valuelooks more like a natural distribution now instead of exponential;
#    - Monthly Premium Auto still resembles an exponential distribution
#    - Number of Open Complaints is much more flat and noes not really resemble normal or exp. distribution anumore.
#    - Number of Policies still looks like exponential
#    - Total Claim Amount is now a mix of exponential & normal distribution

# Additionally:
#    - Income started to have a half-normal (single bell, excluding zeros) shape;
#    - Mopnths since policy inception adopted a nice exponential shape, from undefined before
```

```python
#3 continued
# Do logn vs log10 make any difference in the shape?

numerical = data.select_dtypes(include='number')

def log10_transfom_clean(x):
    if x!=0:
        return np.log10(x)
    else:
        return x # We are returning NaNs so that we can replace them later

for column in numerical.columns:         
    numerical[column] = numerical[column].apply(log10_transfom_clean)
    
def plots():
    for column in numerical.columns:
        sns.displot(numerical[column], bins = 30)
        plt.show()
        
plots()

# No, it just changes the min/max value thresholds
```
